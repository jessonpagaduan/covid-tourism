```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
url_raw = 'https://raw.githubusercontent.com/jessonpagaduan/covid-tourism/master/travel-bans-asof10Aug-python.csv'
```

```python
dates = pd.date_range('20200301', '20200810')
```

```python
dates
```

```python
dates = dates.strftime("%m/%d/%y")
```

```python
dates
```

```python
df = pd.read_csv(url_raw, index_col=0)
```

```python
df.columns = dates
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.index
```

```python
fig, ax = plt.subplots(figsize=(10, 20))
ax = sns.heatmap(df, cmap=sns.diverging_palette(10, 240, n=2), linewidth=0.004, 
                 xticklabels=31, cbar=False 
#                  cbar_kws={'ticks': [0, 1],
#                            'orientation': 'horizontal',
#                            'label': 'Closed border             Open border'}
                )
ax.set_title('COVID-19 travel bans: Which countries have reopened borders?')
ax.set_ylabel('')
ax.annotate('Sources: IATA; New York Times; Al Jazeera; and national sources.', (0.1, 0.05), xycoords='figure fraction')
plt.show()
```
