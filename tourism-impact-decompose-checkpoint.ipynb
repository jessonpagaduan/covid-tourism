```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/jessonpagaduan/covid-tourism/master/tourism-impact-2020-2021.csv'
```

```python
df = pd.read_csv(url)
```

```python
df
```

```python
df.describe()
```

```python
df.plot(kind='box')
```

```python
ent_months_better = 10.935
ent_months_base = 13.935
ent_months_worse = 16.935
ent_months_2020 = 9
```

```python
df = df.assign(dec_better_2020 = lambda x: x.totdec_better * ent_months_2020 / 12)
df = df.assign(dec_better_2021 = lambda x: x.totdec_better * 
               (ent_months_better - ent_months_2020) / 12)
df = df.assign(dec_base_2020 = lambda x: x.totdec_base * ent_months_2020 / 12)
df = df.assign(dec_base_2021 = lambda x: x.totdec_base * 
               (ent_months_base - ent_months_2020) / 12)
df = df.assign(dec_worse_2020 = lambda x: x.totdec_worse * ent_months_2020 / 12)
df = df.assign(dec_worse_2021 = lambda x: x.totdec_worse * 
               (ent_months_worse - ent_months_2020) / 12)
```

```python
df
```

```python
df.to_excel('tourism-impact-2020-21.xlsx')
```

```python
df.plot(kind='box')
```

```python
df.set_index('code')[['dec_better_2020', 'dec_better_2021']].plot(kind='bar')
```

```python
df.set_index('code')[['dec_base_2020', 'dec_base_2021']].plot(kind='bar')
```

```python
df.set_index('code')[['dec_worse_2020', 'dec_worse_2021']].plot(kind='bar')
```

```python
better_sorted = df.set_index('code')[['dec_better_2020', 'dec_better_2021']].sort_values(by='dec_better_2020')
base_sorted = df.set_index('code')[['dec_base_2020', 'dec_base_2021']].sort_values(by='dec_base_2020')
worse_sorted = df.set_index('code')[['dec_worse_2020', 'dec_worse_2021']].sort_values(by='dec_worse_2020')
```

```python
labels = better_sorted.iloc[:15, 0].index
better_2020 = better_sorted.iloc[:15, 0]
better_2021 = better_sorted.iloc[:15, 1]
base_2020 = base_sorted.iloc[:15, 0]
base_2021 = base_sorted.iloc[:15, 1]
worse_2020 = worse_sorted.iloc[:15, 0]
worse_2021 = worse_sorted.iloc[:15, 1]
x = np.arange(len(labels))
width = 0.35

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(30, 8))

axs[0].bar(x - width/2, better_2020, width, label='2020', color='red')
axs[0].bar(x + width/2, better_2021, width, label='2021', color='blue')
axs[0].set_xticks(x)
axs[0].set_xticklabels(labels, rotation=90)
axs[0].set_ylabel('% of GDP')
axs[0].set_ylim([-50, 0])
axs[0].set_title('Better scenario (6 months of travel bans)')
axs[0].legend()

axs[1].bar(x - width/2, base_2020, width, label='2020', color='red')
axs[1].bar(x + width/2, base_2021, width, label='2021', color='blue')
axs[1].set_xticks(x)
axs[1].set_xticklabels(labels, rotation=90)
axs[1].set_ylabel('% of GDP')
axs[1].set_ylim([-50, 0])
axs[1].set_title('Baseline scenario (9 months of travel bans)')

axs[2].bar(x - width/2, worse_2020, width, label='2020', color='red')
axs[2].bar(x + width/2, worse_2021, width, label='2021', color='blue')
axs[2].set_xticks(x)
axs[2].set_xticklabels(labels, rotation=90)
axs[2].set_ylabel('% of GDP')
axs[2].set_ylim([-50, 0])
axs[2].set_title('Worse scenario (12 months of travel bans)')

plt.show()
```

```python
labels = better_sorted.iloc[:15, 0].index
better_2020 = better_sorted.iloc[:15, 0]
better_2021 = better_sorted.iloc[:15, 1]
base_2020 = base_sorted.iloc[:15, 0]
base_2021 = base_sorted.iloc[:15, 1]
worse_2020 = worse_sorted.iloc[:15, 0]
worse_2021 = worse_sorted.iloc[:15, 1]
x = np.arange(len(labels))
width = 0.35

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(30, 8))

axs[0].stem(x - width/2, better_2020, width, label='2020', linefmt='r-')
axs[0].stem(x + width/2, better_2021, width, label='2021', 
            linefmt='b-', markerfmt='bd')
axs[0].set_xticks(x)
axs[0].set_xticklabels(labels, rotation=90, fontsize=15)
axs[0].set_ylabel('% of GDP')
axs[0].set_ylim([-50, 0])
axs[0].set_title('Better scenario (6 months of travel bans)', fontsize=15)
axs[0].legend(loc='lower center', ncol=2)

axs[1].stem(x - width/2, base_2020, width, label='2020', linefmt='r-')
axs[1].stem(x + width/2, base_2021, width, label='2021', 
            linefmt='b-', markerfmt='bd')
axs[1].set_xticks(x)
axs[1].set_xticklabels(labels, rotation=90, fontsize=15)
axs[1].set_ylabel('% of GDP')
axs[1].set_ylim([-50, 0])
axs[1].set_title('Baseline scenario (9 months of travel bans)', fontsize=15)
axs[1].legend(loc='lower center', ncol=2)

axs[2].stem(x - width/2, worse_2020, width, label='2020', linefmt='r-')
axs[2].stem(x + width/2, worse_2021, width, label='2021', 
            linefmt='b-', markerfmt='bd')
axs[2].set_xticks(x)
axs[2].set_xticklabels(labels, rotation=90, fontsize=15)
axs[2].set_ylabel('% of GDP')
axs[2].set_ylim([-50, 0])
axs[2].set_title('Worse scenario (12 months of travel bans)', fontsize=15)
axs[2].legend(loc='lower center', ncol=2)

plt.show()
```

```python
df_all = df.set_index('code').join(df_old.set_index('code'), rsuffix='old')
```

```python
df_all.drop(columns='regionold', inplace=True)
```

```python
df_scatter_1 = df_all[['dec_base_2020', 'dec_6months_2020', 'dec_base_2021', 'dec_6months_2021']] * -1
```

```python
df_scatter_1
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(10, 8))

countries_mask = ['MDV', 'COK', 'PLW', 'VUT', 'WSM', 'GEO', 'FJI', 'KHM']

ax.scatter(x=df_scatter_1.dec_6months_2020, y=df_scatter_1.dec_base_2020, color='b', s=100)

for line in range(0, df_scatter_1.shape[0]):
    if df_scatter_1.index[line] in countries_mask:
        ax.text(df_scatter_1.dec_6months_2020[line]+1, df_scatter_1.dec_base_2020[line], 
                df_scatter_1.index[line], horizontalalignment='left', 
                size=12, color='black')
        


ax.set_xlim([0, 40])
ax.set_ylim([0, 40])
x = np.linspace(*ax.get_xlim())
ax.plot(x, x, '--')
ax.set(ylabel='Latest baseline estimates, % of GDP', 
       xlabel='Old estimates, % of GDP',
       title='Impact of COVID-19 on international tourism receipts in 2020\n\
       New vs Old estimates')


plt.show()
```
