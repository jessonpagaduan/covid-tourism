```python
# Import pandas, matplotlib, seaborn, numpy, and plotly
import pandas as pd
import plotly_express as px
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

# Read CSV
df = pd.read_csv('flightradar24-flights-data.csv', parse_dates=[0])

# Melt dataframe for plotting
df = df.melt(id_vars='DateTime')

# Create an interactive line chart
fig = px.line(data_frame=df, x='DateTime', y='value', color='variable', 
              color_discrete_sequence=['red', 'red', 'blue', 'blue'],
              line_dash='variable',
              line_dash_sequence=['solid', 'dot', 'solid', 'dot'],
              labels={'variable': ''},
              title='Number of commercial flights tracked by Flightradar24 per day (UTC)<br>2019 vs 2020',
              template='plotly_white')

# Update layout
fig.update_layout({'xaxis': {'title': '',
                             'showgrid': False},
                   'yaxis': {'title': '',
                             'showgrid': False}})

fig.show()
```
